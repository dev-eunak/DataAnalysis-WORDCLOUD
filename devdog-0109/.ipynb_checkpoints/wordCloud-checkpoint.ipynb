### 1. 모듈 및 폰트 설정

```python
import re 
from os import path
import numpy as np
from PIL import Image
import pandas as pd
from konlpy.tag import Kkma
from wordcloud import WordCloud, STOPWORDS
import matplotlib.pyplot as plt

from matplotlib import font_manager, rc

# 폰트 지정
font_name = font_manager.FontProperties(fname="c:/Windows/Fonts/malgun.ttf").get_name()
rc('font', family=font_name)
kkma = Kkma()
```

### 2. 텍스트 가져오기

```python
with open('devdog.txt', 'r', encoding='utf-8') as f:
    read_data = f.readlines()
```

```python
read_data[0:10]
```

```python
prepro = []
for i in read_data:
    prepro.append(i.rstrip('\n'))
```

### 3. 명사 분석

```python
tokens = []
for i in prepro:
    tokens += kkma.nouns(str(i))
```

```python
tokens[:20]
```

### 4. 전처리

```python
for i in range(len(tokens)):
    tokens[i] = re.sub(".?[0-9]+.?","", str(tokens[i]))
    if len(tokens[i]) == 1:
        tokens[i] = ""
```

```python
tokens[:20]
```

```python
for i in tokens:
    try:
        tokens.remove("")
    except ValueError:
        break
```

```python
tokens[:20]
```

### 5. 제거할 단어 정하기

```python
stopwords = set(STOPWORDS)
stopwords = stopwords | {'브랜드', '브랜드평판'} # <- 여기에 추가
```

```python
wordcloud = WordCloud(
    width = 500,
    height = 500,
    stopwords = stopwords,
    font_path = 'C:/Windows/Fonts/malgun.ttf',
    max_words = 300,
    background_color = 'white'
).generate(' '.join(tokens))
```

### 6. 워드클라우드 실행

```python
fig = plt.figure(figsize = (8, 8))
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis("off")
plt.title("'데이터 분석' 검색 워드클라우드", fontdict={'fontsize': 20})
plt.show()
```

### 6-1. 워드클라우드 이미지 입히기

```python
mask_img = np.array(Image.open('mask.png'))
```

```python
wordcloud = WordCloud(
    width = 500,
    height = 500,
    stopwords = stopwords,
    font_path = 'C:/Windows/Fonts/malgun.ttf',
    max_words = 300,
    background_color = 'white',
    mask = mask_img
).generate(' '.join(tokens))
```

```python
fig = plt.figure(figsize = (8, 8))
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis("off")
plt.title("'데이터 분석' 검색 워드클라우드", fontdict={'fontsize': 20})
plt.show()
```

```python
# 이미지 저장
wordcloud.to_file("wordcloud.png")
```
